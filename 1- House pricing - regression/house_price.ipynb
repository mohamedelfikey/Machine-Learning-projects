### Detect & Handling outlayers

```python
fig, axes = plt.subplots(nrows=8, ncols=2, figsize=(15, 35))  
    
for i, col in enumerate(df.columns):
    ax = axes[i // 2, i % 2]  
    sns.boxplot(data=df.iloc[:,:], x=col, ax=ax)
```

```python
def handling_outlayers(df):
    for col in df.columns:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1

        upper_limit=Q3+IQR*1.5
        lower_limit=Q1-IQR*1.5

        df[col]=np.where(df[col]<lower_limit,lower_limit,df[col])
        df[col]=np.where(df[col]>upper_limit,upper_limit,df[col])
    return df

df=handling_outlayers(df)
```

### Split data

```python
x=df.drop(columns=['price'])
y=df['price']
print (x.shape)
print (y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print (f'x_train shape : {x_train.shape}')
print (f'y_train shape : {y_train.shape}')
print (f'x_test shape : {x_test.shape}')
print (f'y_test shape : {y_test.shape}')
```

### Data Scalling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

## Model

### 1- Simple linear regression

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train,y_train)
preds=model.predict(x_test)
```

#### Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print (f"mae = {mean_absolute_error(y_test,preds)}")
print (f"mse = {mean_squared_error(y_test,preds)}")
print (f"r2 score = {r2_score(y_test,preds)}")
```

### 2- polynomial linear regression

```python
from sklearn.preprocessing import PolynomialFeatures

for d in range(2,5):
    # polynomial features
    poly=PolynomialFeatures(d)
    x_poly=poly.fit_transform(x)
    # scalling x
    x_scaled=scaler.fit_transform(x_poly)
    # split data
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=42)
    # build model
    model.fit(x_train,y_train)
    train_pred=model.predict(x_train)
    print ("degree =",d)
    print ("For training:")
    print(f"mse: {mean_squared_error(y_train, train_pred)}, mae: {mean_absolute_error(y_train, train_pred)}, r2: {r2_score(y_train, train_pred)}")
    test_pred=model.predict(x_test)
    print ("For testing:")
    print(f"mse: {mean_squared_error(y_test, test_pred)}, mae: {mean_absolute_error(y_test, test_pred)}, r2: {r2_score(y_test, test_pred)}")
```

```python
poly=PolynomialFeatures(3)
x_poly=poly.fit_transform(x)
# scalling x
x_scaled=scaler.fit_transform(x_poly)
# split data
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=42)
# build model
model.fit(x_train,y_train)
```

```python
preds=model.predict(x_test)
print(f"mse: {mean_squared_error(y_test, preds)}, mae: {mean_absolute_error(y_test, preds)}, r2: {r2_score(y_test, preds)}")
```

### 3- SVR

```python
from sklearn.svm import SVR

kernel=['linear','poly','rbf']
for k in kernel: 
    model=SVR(C=1.0,kernel=k,epsilon=0.1)
    model.fit(x_train,y_train)

    preds=model.predict(x_test)
    # evaluation
    print (f"for {k}")
    print(f"mse: {mean_squared_error(y_test, preds)}, mae: {mean_absolute_error(y_test, preds)}, r2: {r2_score(y_test, preds)}")
```

```python
C=[1.0,0.5,0.05,0.01,0.005,0.001]
for c in C:
    model=SVR(C=c,kernel='linear',epsilon=0.1)
    model.fit(x_train,y_train)
    preds=model.predict(x_test)
    # evaluation
    print(f"mse: {mean_squared_error(y_test, preds)}, mae: {mean_absolute_error(y_test, preds)}, r2: {r2_score(y_test, preds)}")
```

```python
eps=[0.01,0.1,0.5,1.0]
for e in eps:
    model=SVR(C=1.0,kernel='linear',epsilon=e)
    model.fit(x_train,y_train)
    preds=model.predict(x_test)
    # evaluation
    print(f"mse: {mean_squared_error(y_test, preds)}, mae: {mean_absolute_error(y_test, preds)}, r2: {r2_score(y_test, preds)}")
```

### 4- XGBRegressor

```python
from xgboost import XGBRegressor
model=XGBRegressor()
model.fit(x_train,y_train)

pred_train=model.predict(x_train)
pred_test=model.predict(x_test)
```

#### Model Evaluation

```python
print ("for train:")
print (f"mae : {mean_absolute_error(y_train,pred_train)}")
print (f"r2 socre {r2_score(y_train,pred_train)}")

print ("for test:")
print (f"mae : {mean_absolute_error(y_test,pred_test)}")
print (f"r2 socre {r2_score(y_test,pred_test)}")
```

### 5- Neural network & Deep learning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

model=Sequential()
model.add(Dense(1028 , activation='relu',input_dim=x_train.shape[1]))
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='adam', loss='mean_squared_error')
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(x_train, y_train, epochs=30, batch_size=50, validation_split=0.2, callbacks=[early_stopping])
```

```python
pred=model.predict(x_train)
print (f"r2 score for train: {r2_score(y_train,pred)}")
```

```python
pred=model.predict(x_test)
print (f"r2 score for test: {r2_score(y_test,pred)}")
```
