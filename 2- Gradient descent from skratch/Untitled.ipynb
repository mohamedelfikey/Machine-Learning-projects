## Gradient descent from skratch

```python
import numpy as np

def gradient_descent(x,y,learning_rate=0.01,iterations=10000):
    theta_0 = theta_1 = 0
    m = len(x)

    for i in range(iterations):
        y_hat = theta_0 + theta_1 * y
        cost = 1/2* m * (sum([value** 2 for value in (y-y_hat)]))
        der_theta_0 = 1/m * (sum(y_hat-y))
        der_theta_1 = 1/m * ( x * sum(y_hat - y))
        theta_0 = theta_0 - learning_rate * der_theta_0
        theta_1 = theta_1 - learning_rate * der_theta_1
        print(f"theta_0 = {theta_0} , theta_1 = {theta_1} , cost = {cost} in iteration : {i}")
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

gradient_descent(x,y,iterations=100)
```
