# Normal equation from skratch

### Import basic library and read data

### Apply normal equation

##### Split data to x,y

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
print (f"shape of x: {x.shape}")
print (f"shape of y: {y.shape}")
```

```python
x_array=x.to_numpy()
x_array
```

```python
n=len(x_array)
x_bias=np.ones((n,1))
x_bias
```

```python
new_x=np.append(x_bias,x_array,axis=1)
new_x
```

##### apply normal equation

```python
new_x_t=new_x.T  # x_t -> x_transpose
new_x_t_dot_new_x=new_x_t.dot(new_x) # x_t . x
temp1=np.linalg.inv(new_x_t_dot_new_x) # (x_t * x) INVERSE
temp2=new_x_t.dot(y) # x_t . y
theta_hat=temp1.dot(temp2) # (x_t.x)INV .x_t .y
theta_hat
```

```python
theta_0=theta_hat[0]
theta_1=theta_hat[1]
theta_2=theta_hat[2]
theta_3=theta_hat[3]
```

### Make predication

```python
x1=x.iloc[:,0]
x2=x.iloc[:,1]
x3=x.iloc[:,2]
print (f"shape of x1: {x1.shape}")
print (f"shape of x1: {x2.shape}")
print (f"shape of x1: {x3.shape}")
```

```python
pred=theta_0+theta_1*x1+theta_2*x2+theta_3*x3
pred
```

### Evaluate model

```python
from sklearn.metrics import mean_absolute_error, r2_score

print (f"mse = {mean_absolute_error(y,pred)}")
print (f"r2 score = {r2_score(y,pred)}")
```

### predication values

```python
def pred_sales(youtube,facebook,news_paper):
    return (theta_0 + theta_1*youtube + theta_2*facebook + theta_3*news_paper)

youtube=84
facebook=19
news_paper=20
print (f"Sales = {pred_sales(youtube,facebook,news_paper)}" )
```
